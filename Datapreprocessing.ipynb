```python
#Installing the required libraries
import pandas as pd, numpy as np # For Data Manipulation
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder # For Encoding Categorical Data [Nominal | Ordinal]
from sklearn.preprocessing import OneHotEncoder # For Creating Dummy Variables of Categorical Data [Nominal]
from sklearn.impute import SimpleImputer, KNNImputer # For Imputation of Missing Data
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler # For Rescaling Data
from sklearn.model_selection import train_test_split # For Splitting Data into Training & Testing Sets
```

```python
#Loading the dataset
data=pd.read_csv('/content/healthcare-dataset-stroke-data.csv')
```

# **1.Treatment of Missing Data [Null, NaN, None]**

```python
#Analysing the shape of the dataset & displaying the first 5 rows
data.shape,data.head(5)
```

```python
#Details of the dataset
data.info()
```

```python
#Converting String to Boolean Type
data['ever_married'] = data['ever_married'].replace({'Yes': 1, 'No': 0})
```

**1.1. Missing Data Information**

```python
#Checking for missing values columnwise
data_missing=data.isna().sum()
#Record-wise Missing Data Information (Top 5)
record_missing_data = data.isna().sum(axis=1).sort_values(ascending=False).head(5)
data_missing,record_missing_data
```

```python
# Data Bifurcation
# Non-Categorical Data
df_noncat = data[['id','age','hypertension','heart_disease','ever_married','avg_glucose_level','bmi','stroke']]
# Categorical Data [Nominal | Ordinal]
df_cat = data[['gender', 'work_type','Residence_type', 'smoking_status']]
```

**1.2. Impute Missing Categorical Data [Nominal | Ordinal] using Descriptive Statistics : Central Tendency**

```python
# Strategy = median [When Odd Number of Categories Exists]
si_cat = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
si_cat_fit = si_cat.fit_transform(df_cat)
# Missing Categorical Data Imputed Subset
df_cat_mdi = pd.DataFrame(si_cat_fit, columns=df_cat.columns)
df_cat_mdi , df_cat_mdi.info()
```

**1.3. Impute Missing Non-Categorical Data using Descriptive Statistics : Central Tendency**

```python
# Strategy : mean | median | most_frequent | constant
si_noncat = SimpleImputer(missing_values=np.nan, strategy='mean')
si_noncat_fit = si_noncat.fit_transform(df_noncat)
# Missing Non-Categorical Data Imputed Subset using Simple Imputer
df_noncat_mdi_si = pd.DataFrame(si_noncat_fit, columns=df_noncat.columns)
df_noncat_mdi_si,  df_noncat_mdi_si.info()
```

**1.4. Impute Missing Non-Categorical Data using K-Nearest Neighbors**

```python
# weights : uniform | distance | {User Defined}
ki = KNNImputer(n_neighbors=5, weights='uniform')
ki_fit = ki.fit_transform(df_noncat)
# Missing Non-Categorical Data Imputed Subset using KNN Imputer
df_noncat_mdi_ki = pd.DataFrame(ki_fit, columns=df_noncat.columns)
df_noncat_mdi_ki, df_noncat_mdi_ki.info()
```

**1.5. Missing Data Exclusion [MCAR | MAR (> 50%)]**

```python
# Dataset Used : df_cat_mdi | df_noncat_mdi_si | df_noncat_mdi_ki

# Excluding Empty Records (If Any)
df_cat_mdi.dropna(axis=0, how='all', inplace=True) # Categorical Data Subset
# Non-Categorical Data Subset
df_noncat_mdi_si.dropna(axis=0, how='all', inplace=True)
# Non-Categorical Data Subset
df_noncat_mdi_ki.dropna(axis=0, how='all', inplace=True)
```

```python
# Excluding Empty Variables (If Any)
df_cat_mdi.dropna(axis=1, how='all', inplace=True) # Categorical Data Subset
# Non-Categorical Data Subset
df_noncat_mdi_si.dropna(axis=1, how='all', inplace=True)
# Non-Categorical Data Subset
df_noncat_mdi_ki.dropna(axis=1, how='all', inplace=True)
```

```python
# Missing Categorical Treated Dataset
df_cat_mdt = df_cat_mdi.copy()
# Missing Non-Categorical Treated Dataset
df_noncat_mdt = df_noncat_mdi_si.copy()
# Missing Non-Categorical Treated Dataset
df_noncat_mdt = df_noncat_mdi_ki.copy()
```

# **2. Numeric Coding of Data**

***2.1. Numeric Encoding of Categorical Data [Nominal & Ordinal] ***

```python
df_cat_mdt_code = df_cat_mdt.copy()
# Using Pandas (Inferior)
df_cat_mdt_code_pd = df_cat_mdt_code.astype('category')
df_cat_mdt_code_pd['gender_code'] = df_cat_mdt_code_pd['gender'].cat.codes
df_cat_mdt_code_pd['work_type_code'] = df_cat_mdt_code_pd['work_type'].cat.codes
df_cat_mdt_code_pd['Residence_type_code'] = df_cat_mdt_code_pd['Residence_type'].cat.codes
df_cat_mdt_code_pd['smoking_status_code'] = df_cat_mdt_code_pd['smoking_status'].cat.codes
# (Missing Data Treated) Numeric Coded Categorical Dataset using Pandas
df_cat_mdt_code_pd
```

```python
# Using Scikit Learn : Label Encoder (Inferior)
le = LabelEncoder()
le_fit1 = le.fit_transform(df_cat_mdt_code['gender'])
le_fit2 = le.fit_transform(df_cat_mdt_code['work_type'])
le_fit3 = le.fit_transform(df_cat_mdt_code['Residence_type'])
le_fit4 = le.fit_transform(df_cat_mdt_code['smoking_status'])
df_gen_code = pd.DataFrame(le_fit1, columns=['gender_code']); df_gen_code
df_work_code = pd.DataFrame(le_fit2, columns=['work_type_code']); df_work_code
df_resi_code = pd.DataFrame(le_fit3, columns=['Residence_type_code']); df_resi_code
df_smok_code = pd.DataFrame(le_fit4, columns=['smoking_status_code']); df_smok_code
df_gen_work_code = df_gen_code.join(df_work_code); df_gen_work_code
df_gen_work_resi_code = df_gen_work_code.join(df_resi_code); df_gen_work_resi_code
df_gen_work_resi_smok_code = df_gen_work_resi_code.join(df_smok_code); df_gen_work_resi_smok_code
# (Missing Data Treated) Numeric Coded Categorical Dataset using Scikit Learn Label Encoder
df_cat_mdt_code_le = df_cat_mdt_code.join(df_gen_work_resi_smok_code); df_cat_mdt_code_le
df_cat_mdt_code_le = pd.merge(df_cat_mdt_code, df_gen_work_resi_smok_code, left_index=True, right_index=True); df_cat_mdt_code_le
```

```python
# Using Scikit Learn : Ordinal Encoder (Superior)
oe = OrdinalEncoder()
oe_fit = oe.fit_transform(df_cat_mdt_code)
df_cat_code_oe = pd.DataFrame(oe_fit, columns=['gender_code', 'work_type_code','Residence_type_code','smoking_status_code']); df_cat_code_oe
# (Missing Data Treated) Numeric Coded Categorical Dataset using Scikit Learn Ordinal Encoder
df_cat_mdt_code_oe = df_cat_mdt_code.join(df_cat_code_oe); df_cat_mdt_code_oe
df_cat_mdt_code_oe = pd.merge(df_cat_mdt_code, df_cat_code_oe, left_index=True, right_index=True); df_cat_mdt_code_oe
```

***2.2. Creation of Dummy Variables from Categorical Data ***

```python
df_cat_mdt_code = df_cat_mdt.copy()
# Dummy Variable : gender
# Using Pandas (Superior)
df_mdt_gen_dummy_pd = pd.get_dummies(df_cat_mdt_code[['gender']]); df_mdt_gen_dummy_pd # (Missing Data Treated) Dummy Dataset using Pandas
df_mdt_cat_dummy_pd = pd.get_dummies(df_cat_mdt_code, columns = ['gender', 'work_type','Residence_type','smoking_status']); df_mdt_cat_dummy_pd
```

```python
# Using Scikit Learn (Inferior)
gen_cat = df_cat_mdt_code['gender'].unique(); gen_cat
ohe = OneHotEncoder(handle_unknown='ignore')
ohe_fit = ohe.fit_transform(df_cat_mdt_code[['gender']])
# (Missing Data Treated) Dummy Dataset using Scikit Learn One Hot Encoder
df_mdt_gen_dummy_ohe = pd.DataFrame(ohe_fit.toarray(), columns='gender_'+gen_cat); df_mdt_gen_dummy_ohe
```

# **3. Data Transformation & Rescaling [Treatment of Outliers]**

**3.1. Standardization**

```python
ss = StandardScaler()
# Scaling Variable : bmi
ss_fit = ss.fit_transform(df_noncat_mdt[['bmi']])
df_noncat_std = pd.DataFrame(ss_fit, columns=['bmi_std']); df_noncat_std
#df_noncat_std = pd.DataFrame(ss_fit, columns=df_noncat_mdt.columns+'_std'); df_noncat_std
# (Missing Data Treated) Standardized Non-Categorical Dataset using Scikit Learn Standard Scaler
df_noncat_mdt_std = df_noncat_mdt.join(df_noncat_std); df_noncat_mdt_std
df_noncat_mdt_std = pd.merge(df_noncat_mdt, df_noncat_std, left_index=True, right_index=True); df_noncat_mdt_std
```

**3.2. Normalization : Min-Max Scaling**

```python
mms = MinMaxScaler()
mms_fit = mms.fit_transform(df_noncat_mdt[['bmi']])
df_noncat_minmax_norm = pd.DataFrame(mms_fit, columns=['bmi_mmnorm']); df_noncat_minmax_norm
#df_noncat_minmax_norm = pd.DataFrame(mms_fit, columns=df_noncat_mdt.columns+'_mmnorm'); df_noncat_minmax_norm
# (Missing Data Treated) Normalized Non-Categorical Dataset using Sikit Learn Min-Max Scaler
df_noncat_mdt_mmn = df_noncat_mdt.join(df_noncat_minmax_norm); df_noncat_mdt_mmn
df_noncat_mdt_mmn = pd.merge(df_noncat_mdt, df_noncat_minmax_norm, left_index=True, right_index=True); df_noncat_mdt_mmn
```

***3.3. Normalization : Robust Scaling ***

```python
# quantile_range=(25.0, 75.0) - Default Range
rs = RobustScaler(quantile_range=(10.0, 90.0))
rs_fit = rs.fit_transform(df_noncat_mdt[['bmi']])
df_noncat_robust_norm = pd.DataFrame(rs_fit, columns=['bmi_robnorm']); df_noncat_robust_norm
#df_noncat_robust_norm = pd.DataFrame(rs_fit, columns=df_noncat_mdt.columns+'_robnorm'); df_noncat_robust_norm
# (Missing Data Treated) Normalized Non-Categorical Dataset using Scikit Learn Robust Scaler
df_noncat_mdt_rn = df_noncat_mdt.join(df_noncat_robust_norm); df_noncat_mdt_rn
df_noncat_mdt_rn = pd.merge(df_noncat_mdt, df_noncat_robust_norm, left_index=True, right_index=True); df_noncat_mdt_rn
```

**3.4. Log Transformation**

```python
df_noncat_mdt_lt = df_noncat_mdt.copy()
# (Missing Data Treated) Log Tranformed Non-Categorical Dataset
df_noncat_mdt_lt['ln_bmi'] = np.log(df_noncat_mdt_lt['bmi']); df_noncat_mdt_lt
df_noncat_mdt_lt['log10_bmi'] = np.log10(df_noncat_mdt_lt['bmi']); df_noncat_mdt_lt
```

```python
# Pre-Processed Categorical Data Subset
df_cat_ppd = df_cat_mdt_code_oe.copy(); df_cat_ppd # Preferred Data Subset
# Pre-Processed Non-Categorical Data Subset
df_noncat_ppd = df_noncat_mdt_mmn.copy(); df_noncat_ppd # Preferred Data Subset
# Pre-Processed Dataset
df_ppd = df_cat_ppd.join(df_noncat_ppd); df_ppd # Pre-Processed Dataset
df_ppd = pd.merge(df_cat_ppd, df_noncat_ppd, left_index=True, right_index=True); df_ppd
```

# **4. Data Bifurcation [Training & Testing Datasets]**

```python
train_df, test_df = train_test_split(df_ppd, test_size=0.25, random_state=1234)
# Training Dataset
train_df
# Testing Dataset
test_df, train_df
```
